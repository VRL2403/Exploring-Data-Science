```python
import numpy as np
import pandas as pd
import os
```

#### Sample dataset (movies dataset) taken from Kaggle.

```python
master = pd.read_csv('movies.csv')
```

```python
# Create DataFrame to store stats of different file format (file size and read, write time).
df_results = pd.DataFrame(columns=['method', 'file_size', 'write_time', 'read_time'])
display(df_results)
```

```python
# Function to store results in DataFrame.
def add_result(df_results, method, file_size, write_time, read_time):
    row = {
        'method': method,
        'file_size': file_size,
        'write_time': write_time,
        'read_time':  read_time
    }
    return df_results.append(pd.Series(row), ignore_index=True)
```

```python
filename = 'Final'
```

```python
# Write File.
result_save = %timeit -n5 -r5 -o master.to_csv(filename + '_csv.csv')

# Size of File
filesize = os.path.getsize(filename + '_csv.csv') / 1024**2

# Read File.
result_read = %timeit -n5 -r5 -o pd.read_csv(filename + '_csv.csv')

# Save the results.
df_results = add_result(df_results,
                        'CSV',
                        filesize,
                        result_save.average,
                        result_read.average)
df_results
```
